# MCQA: eval by category

```python
import pandas as pd
import json

from datasets import load_dataset, DatasetDict
from utils import *
```

```python
df = pd.DataFrame(read_jsonl("datasets/MCQA_eval.jsonl"))
df.head()
```

```python
def create_subject_jsonl(sub):
    sub_df = df[df["subject"] == sub]
    name = 'eval_' + sub + '.jsonl'
    sub_df.to_json(name, orient='records', lines=True)

for subject in df["subject"].unique():
    create_subject_jsonl(subject)
```

```python
for subject in df["subject"].unique():
    print(f'{subject} : {len(df[df["subject"] == subject])}')
```
